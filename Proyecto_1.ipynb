**Ejercicio N°1:**

Escribir un programa que pregunte al usuario por las ventas de los 
últimos 5 días, y muestre por pantalla una serie con los datos de 
las ventas por producto: Nro de unidades y precio unitario, 
indexada por los días, antes y después de aplicarles un descuento 
del 5%.

```python
import numpy as np
import pandas as pd
```

```python
#Entrada de datos

l_ventas = [ ["Mouse", 0, 7.0, 0],
             ["Teclado", 0, 10.0, 0],
             ["Monitor", 0, 70.0, 0],
             ["Parlante", 0, 12.0, 0],
             ["Router", 0, 53.0, 0]     ]

np.random.seed(123)

df_ventas = pd.DataFrame( 
    l_ventas,
    index = ["20220410", "20220411", "20220412", "20220413", "20220414"], #df con etiqueta en el index
    columns=["PRODUCTO", "CANTIDAD", "PRECIO_USD", "IMPORTE"]  )

#print(df_ventas.head())


for i in range(len(df_ventas)):
  fecha = df_ventas.index[i]; 
  producto = df_ventas.iat[i,0]
  cant = int(input (f"Registre la cantidad de ventas del producto: {producto}, en fecha: {fecha}   --> "))
  if cant >  0:
    df_ventas.iat[i,1] = cant
  else:
    print ("Debe registrar un número válido.")
    break
print("_______________________________________________________________________")
df_ventas.iloc[:,0:3]
```

```python
# Cálculo de importe total
df_ventas["IMPORTE"] = df_ventas.CANTIDAD * df_ventas.PRECIO_USD
print("VENTAS")
print("___________________________________________________")
df_ventas
```

```python
# Cálculo de importe con descuento del 5%
df_ventas["IMPORTE"] *= 0.95
print("VENTAS CON DESCUENTO DEL 5%")
print("_______________________________________________________")
df_ventas
```

**Ejercicio N°2:**



Escribir un programa que reciba un diccionario con los ingresos 
mensuales de los clientes de una empresa, y devuelva una serie 
con el ingreso mínimo, máximo y media.

```python
import pandas as pd
import numpy as np
```

```python
def registro_Ingresos(dic_ingreso):
  ''' La Función registro_Ingresos --> devolverá una serie con el ingreso mensual: mínimo, máximo y media, de los clientes de una empresa.'''
  serie_ingresos = pd.Series(dic_ingreso)
  serie_calculo = pd.Series( [serie_ingresos.min(), serie_ingresos.max(), serie_ingresos.mean()],
                                 index = ["INGRESO MÍNIMO", "INGRESO MÁXIMO", "INGRESO MEDIO"] )
  return serie_calculo
```

```python
dic_ingresos = {"Juan Perez M.": 8500,
                "Mariana Gonzales C.": 4000,
                "Lorgio Zabala F.": 6520,
                "Pedro Dorado R.": 14000,
                "Roberth García S.": 17500 }
dic_ingresos.items()
registro_Ingresos(dic_ingresos)
```

**Ejercicio N°3:**

Escribir programa que genere y muestre por pantalla un DataFrame 
con los datos de la tabla siguiente:

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses.
5. Obtener el total de gastos de febrero a abril.


```python
import pandas as pd
import numpy as np
```

```python
df_ventas = pd.DataFrame(columns = ["MES", "VENTAS", "GASTOS"])
df_ventas["MES"] = ["Enero", "Febrero", "Marzo", "Abril", "Marzo"]
df_ventas["VENTAS"] = [ 30500, 35600, 28300, 33900, 42500]
df_ventas["GASTOS"] = [ 22000, 23450, 18100, 35700, 32450]

df_ventas
```

```python
# 1 --> Columna 1 = Gasto/Ventas*100.

df_ventas["COLUMNA 1"] = df_ventas.VENTAS / df_ventas.GASTOS * 100
df_ventas
```

```python
# 2 --> Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”, caso contrario imprimir “Meta no Superada”.

def supera_meta(l_ventas):
  ''' La función determina si el monto de la venta superó la meta o no.'''
  resp = list()
  for i in range(len(l_ventas)):
    if l_ventas[i] >= 30000:
      resp.append("Meta Superada")
    else:
      resp.append("Meta no Superada")
  return resp

l_ventas = list(df_ventas.VENTAS)
df_ventas["META"] = pd.Series(supera_meta(l_ventas))
df_ventas
```

```python
# 3 --> Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.

def supera_gastos(l_gastos, l_ventas):
  ''' La función determina si los gastos son superiores a las ventas realizadas.'''
  resp = list()
  for i in range(len(l_gastos)):
    if l_gastos[i] >= l_ventas[i]:
      resp.append(l_gastos[i]-l_ventas[i])
    else:
      resp.append(0)
  return resp

l_gastos = list(df_ventas.GASTOS)
l_ventas = list(df_ventas.VENTAS)
df_ventas["SOBRECOSTO"] = pd.Series(supera_gastos(l_gastos, l_ventas))
df_ventas
```

```python
# 4 --> Obtener el total de las ventas de los últimos 3 meses.

total_trim = df_ventas.loc[2:,"VENTAS"].sum()
print(f"El total de VENTAS de los últimos 3 meses es: {total_trim}")
```

```python
# 5 --> Obtener el total de gastos de febrero a abril.

total_calc = df_ventas.loc[1:3,"GASTOS"].sum()
print(f"El total de GASTOS de febrero a Abril es: {total_calc}")
```
